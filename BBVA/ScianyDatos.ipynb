```python
def escalar_creditos(datos):
    
    datos = datos.drop(columns="OFERTA CRÉDITO EN 2018")
    datos = datos.drop(columns="TDCN")
    
    creditos_aprobados = datos["FH_FORMA CREDITO 2018"]
    creditos_otorgados = datos["Créditos simples otorgados desde 2018"]
    
    datos["FH_FORMA CREDITO 2018"] = datos["FH_FORMA CREDITO 2018"].replace('#N/D', 0)
    
    
    #datos["Créditos simples otorgados desde 2018"] = datos["Créditos simples otorgados desde 2018"].replace(0, random.uniform(0.3, 0.5))
    #datos["Créditos simples otorgados desde 2018"] = datos["Créditos simples otorgados desde 2018"].replace([0,1,2], random.uniform(0.5, 0.7))
    
    n = len(creditos_aprobados)
    
    for i in range(100):
        if creditos_aprobados[i] == "#N/D":
            creditos_aprobados[i] = 0
        else:
            creditos_aprobados[i] = 1
            
        if creditos_otorgados[i] == 0:
            p = random.uniform(0.3, 0.5)
            creditos_otorgados[i] = p
        else:
            p = random.uniform(0.5, 0.7)
            creditos_aprobados[i] = p
            
        # Si no se pone ésto no cambia la p
        p = 1
        
    #datos = datos.rename(columns={"FH_FORMA CREDITO 2018":"Créditos", "Créditos simples otorgados desde 2018":"Salud"})
                
    return datos
```

```python
def escalar_creditos2(datos):
    
    datos = datos.drop(columns="OFERTA CRÉDITO EN 2018")
    datos = datos.drop(columns="TDCN")
    
    creditos_aprobados = datos["FH_FORMA CREDITO 2018"]
    creditos_otorgados = datos["Créditos simples otorgados desde 2018"]
    
    datos["FH_FORMA CREDITO 2018"] = datos["FH_FORMA CREDITO 2018"].replace('#N/D', 0)
    
    #datos["Créditos simples otorgados desde 2018"] = datos["Créditos simples otorgados desde 2018"].replace(0, random.uniform(0.3, 0.5))
    #datos["Créditos simples otorgados desde 2018"] = datos["Créditos simples otorgados desde 2018"].replace([0,1,2], random.uniform(0.5, 0.7))
    
    n = len(creditos_aprobados)
        
    for i in range(len(creditos_aprobados)):
            
        if creditos_otorgados[i] == 0:
            p = random.uniform(0.3, 0.5)
            
        else:
            p = random.uniform(0.5, 0.7)
            datos["Créditos simples otorgados desde 2018"].replace(creditos_otorgados[i], p, inplace=True)
            
        # Si no se pone ésto no cambia la p
        p = 1
        
    #datos = datos.rename(columns={"FH_FORMA CREDITO 2018":"Créditos", "Créditos simples otorgados desde 2018":"Salud"})
                
    return datos
```

```python
def replace_zeros_w_random_normal(DF,label, mu, sigma):
    truth_1 = DF[label] == 0
    random = np.random.normal(mu, sigma, DF.shape[0])
    filt = DF[DF[label] > 0]
    vector_1 = truth_1 * random
    truth_2 = vector_1 == 0
    vector_2 = truth_2 * DF[label]
    DF[label] = np.maximum(vector_1,vector_2)
    return DF
```
