```python
frames = [carpinteria, constructora, sastreria, mecanica, abogados, servicios_medicos]
```

```python
matriz = pd.concat(frames)
```

```python
matriz.fillna(0)
```

```python
def top5(x):
    x.sort(key = abs, reverse = True)
    
    return [x[i] for i in range(5)]
```

```python
top5([5,-4,3,2,1])
```

```python
matriz = matriz.fillna(0)
m, n = matriz.values.shape

M = []

for i in range(m):
    x = [float(matriz.values[i,k]) for k in range(n)]
    M.append(top5(x))
        
M
```

```python
def normalizar(M):
    m,n = M.shape
    M2 = np.zeros((m,n))
    
    sumas = []
    
    for i in range(m):
        sumas.append(sum([M[i,j] for j in range(n) if M[i,j] > 0]))
        
    for i in range(m):
        for j in range(n):
            M2[i,j] = M[i,j]/sumas[i]
            
    return M2
```
